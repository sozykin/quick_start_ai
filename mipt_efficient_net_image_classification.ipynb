Учебный курс "Быстрый старт в искусственный интеллект" для учителей школ.

```python
from tensorflow.keras.applications import EfficientNetB7
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.applications.efficientnet import decode_predictions
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

Загружаем предварительно обученную модель EfficientNetB7

```python
model = EfficientNetB7(weights='imagenet')
```

Прописываем путь к файлу с изображением

```python
img_path = 'plane.jpg'
```

Загружаем изображение в память

EfficientNetB7 рассчитана на изображения размером 224х224

```python
img = image.load_img(img_path, target_size=(600, 600))
```

```python
plt.imshow(img)
plt.show()
```

Выполняем предварительную обработку изображения

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

Запускаем распознавание

```python
preds = model.predict(x)
```

Печатаем ТОП-3 класса с самой большой вероятностью

```python
classes = decode_predictions(preds, top=3)[0]
for cl in classes:
    print(cl[1], cl[2])
```
